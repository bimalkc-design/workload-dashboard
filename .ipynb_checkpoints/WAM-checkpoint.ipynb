```python
import pandas as pd
from datetime import datetime
import os
from openpyxl import Workbook
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side
from openpyxl.utils import get_column_letter

# ============================================================
# DATA DEFINITIONS - DEPARTMENT OF NATURAL SCIENCES
# ============================================================

# Core Faculty Data - Department of Natural Sciences
faculty_data = [
    # Chemistry Core Faculty
    {'id': 'F1', 'name': 'Dr. Jas Raj Subba', 'program': 'Chemistry', 'designation': 'Professor', 
     'specialization': 'Organic Chemistry / Fundamentals', 'core': True},
    {'id': 'F2', 'name': 'Mr. Sangay Wangchuk', 'program': 'Chemistry', 'designation': 'Lecturer', 
     'specialization': 'Physical Chemistry / Research Methodology', 'core': True},
    {'id': 'F3', 'name': 'Mrs. Punam Mafchan', 'program': 'Chemistry', 'designation': 'Lecturer', 
     'specialization': 'Spectroscopic Methods / Applied Chemistry', 'core': True},
    {'id': 'F4', 'name': 'Ms. Kuenzang Choki', 'program': 'Chemistry', 'designation': 'Lecturer', 
     'specialization': 'Organic Synthetic Frameworks', 'core': True},
    {'id': 'F5', 'name': 'Ms. Sangay Yuden', 'program': 'Chemistry', 'designation': 'Lecturer', 
     'specialization': 'Biochemistry Pathways / Service Support', 'core': True},
    {'id': 'F6', 'name': 'Mr. Tashi Dendup', 'program': 'Chemistry', 'designation': 'Lecturer', 
     'specialization': 'Inorganic Structuring & Coordination Labs', 'core': True},
    
    # Physics Core Faculty
    {'id': 'F7', 'name': 'Shacha Thinley', 'program': 'Physics', 'designation': 'Lecturer', 
     'specialization': 'Atomic Structures & Core Theoretical Physics', 'core': True},
    {'id': 'F8', 'name': 'Nachiketa Homchaudhuri', 'program': 'Physics', 'designation': 'Assistant Professor', 
     'specialization': 'Statistical Mechanics & Analytical Dynamics', 'core': True},
    {'id': 'F9', 'name': 'Mon Bahadur Ghalley', 'program': 'Physics', 'designation': 'Lecturer', 
     'specialization': 'Electromagnetism / Quantum Fields', 'core': True},
    {'id': 'F10', 'name': 'Dr. Karma Tenzin', 'program': 'Physics', 'designation': 'Professor', 
     'specialization': 'Computational Modelling & Metrics', 'core': True},
    {'id': 'F11', 'name': 'Rit Wik Sharma', 'program': 'Physics', 'designation': 'Lecturer', 
     'specialization': 'Mathematical Physics Foundations', 'core': True},
    
    # Life Sciences Core Faculty (6 core members)
    {'id': 'F12', 'name': 'Ugyen Dorji Tamang', 'program': 'Life Sciences', 'designation': 'Lecturer', 
     'specialization': 'Fundamentals of Biological Sciences / Labs', 'core': True},
    {'id': 'F13', 'name': 'Karma Wangchuck', 'program': 'Life Sciences', 'designation': 'Assistant Professor', 
     'specialization': 'Embryology & Economic Botany', 'core': True},
    {'id': 'F14', 'name': 'Mohan Singh Rana', 'program': 'Life Sciences', 'designation': 'Lecturer', 
     'specialization': 'Biochemistry / Animal Physiology', 'core': True},
    {'id': 'F15', 'name': 'Tshering Dekar', 'program': 'Life Sciences', 'designation': 'Lecturer', 
     'specialization': 'Developmental Biology / Freshwater Biology', 'core': True},
    {'id': 'F16', 'name': 'Bimal Kumar Chetri', 'program': 'Life Sciences', 'designation': 'Lecturer', 
     'specialization': 'Plant Breeding & Horticulture', 'core': True},
    {'id': 'F17', 'name': 'Sonam Tobgay', 'program': 'Life Sciences', 'designation': 'Lecturer', 
     'specialization': 'Plant Biotechnology & Tissue Culture', 'core': True},
    
    # Additional Faculty
    {'id': 'F18', 'name': 'Dechen Lhendup', 'program': 'Chemistry', 'designation': 'Lecturer', 
     'specialization': 'Mathematics', 'core': False, 'dept': 'DSDA'},
    {'id': 'F19', 'name': 'S. Chitra', 'program': 'Chemistry', 'designation': 'Lecturer', 
     'specialization': 'Academic Skills', 'core': False, 'dept': 'DSDA'},
    {'id': 'F20', 'name': 'DS-Y', 'program': 'Chemistry', 'designation': 'Lecturer', 
     'specialization': 'Python Programming', 'core': False, 'dept': 'DSDA'},
    {'id': 'F21', 'name': 'Paul Raj', 'program': 'Life Sciences', 'designation': 'Service Lecturer', 
     'specialization': 'Foundation Statistics', 'core': False, 'dept': 'DSDA'},
    {'id': 'F22', 'name': 'LS W', 'program': 'Life Sciences', 'designation': 'Lab Instructor', 
     'specialization': 'Chemistry Lab Instruction', 'core': False, 'dept': 'DSDA'},
    {'id': 'F23', 'name': 'LS X', 'program': 'Life Sciences', 'designation': 'Lab Instructor', 
     'specialization': 'Mathematics Lab Instruction', 'core': False, 'dept': 'DSDA'},
    {'id': 'F24', 'name': 'LS Y', 'program': 'Life Sciences', 'designation': 'Language Instructor', 
     'specialization': 'Dzongkha Instruction', 'core': False, 'dept': 'DSDA'},
    {'id': 'F25', 'name': 'LS Z', 'program': 'Life Sciences', 'designation': 'Academic Skills Instructor', 
     'specialization': 'Academic Skills', 'core': False, 'dept': 'DSDA'}
]

# Module Data - Each unique module counted as 1 (even with separate lab)
module_data = [
    # Chemistry Modules
    {'code': 'OCH201', 'name': 'Organic Chemistry I', 'program': 'Chemistry', 'year': '2', 'credits': 12, 
     'theory': 3, 'lab': 3, 'faculty': 'Dr. Jas Raj Subba'},
    {'code': 'FCH101', 'name': 'Fundamentals of Chemistry', 'program': 'Chemistry', 'year': '1', 'credits': 12, 
     'theory': 2, 'lab': 3, 'faculty': 'Dr. Jas Raj Subba'},
    {'code': 'PCH303', 'name': 'Physical Chemistry III', 'program': 'Chemistry', 'year': '3', 'credits': 12, 
     'theory': 3, 'lab': 3, 'faculty': 'Mr. Sangay Wangchuk'},
    {'code': 'RSM301', 'name': 'Research Methods', 'program': 'Chemistry', 'year': '2', 'credits': 12, 
     'theory': 3, 'lab': 1, 'faculty': 'Mr. Sangay Wangchuk'},
    {'code': 'OCH304', 'name': 'Spectroscopic Methods in Chemistry', 'program': 'Chemistry', 'year': '3', 
     'credits': 12, 'theory': 3, 'lab': 1, 'faculty': 'Mrs. Punam Mafchan'},
    {'code': 'PCH201', 'name': 'Physical Chemistry I', 'program': 'Chemistry', 'year': '2', 'credits': 12, 
     'theory': 3, 'lab': 3, 'faculty': 'Mrs. Punam Mafchan'},
    {'code': 'OCH303', 'name': 'Organic Chemistry III', 'program': 'Chemistry', 'year': '3', 'credits': 12, 
     'theory': 3, 'lab': 3, 'faculty': 'Ms. Kuenzang Choki'},
    {'code': 'BCH301', 'name': 'Principles of Biochemistry I', 'program': 'Chemistry', 'year': '3', 'credits': 12, 
     'theory': 3, 'lab': 3, 'faculty': 'Ms. Sangay Yuden'},
    {'code': 'ICH202', 'name': 'Inorganic Chemistry II', 'program': 'Chemistry', 'year': '2', 'credits': 12, 
     'theory': 3, 'lab': 3, 'faculty': 'Mr. Tashi Dendup'},
    {'code': 'ICH203', 'name': 'Inorganic Chemistry III', 'program': 'Chemistry', 'year': '3', 'credits': 12, 
     'theory': 3, 'lab': 3, 'faculty': 'Mr. Tashi Dendup'},
    {'code': 'FMT204', 'name': 'Foundations of Mathematics III', 'program': 'Chemistry', 'year': '2', 'credits': 12, 
     'theory': 5, 'lab': 0, 'faculty': 'Dechen Lhendup'},
    {'code': 'CSP101', 'name': 'Foundations of Python Programming', 'program': 'Chemistry', 'year': '1', 'credits': 12, 
     'theory': 3, 'lab': 2, 'faculty': 'DS-Y'},
    {'code': 'ACS101', 'name': 'Academic Skills', 'program': 'Chemistry', 'year': '1', 'credits': 12, 
     'theory': 4, 'lab': 0, 'faculty': 'S. Chitra'},
    {'code': 'LAC101', 'name': 'Dzongkha Shoeyon Dretshel', 'program': 'Chemistry', 'year': '1', 'credits': 12, 
     'theory': 4, 'lab': 0, 'faculty': ''},
    
    # Physics Modules
    {'code': 'MPH303', 'name': 'Atomic Physics', 'program': 'Physics', 'year': '3', 'credits': 12, 
     'theory': 3, 'lab': 3, 'faculty': 'Shacha Thinley'},
    {'code': 'PHY101', 'name': 'Fundamentals of Physics (Chemistry)', 'program': 'Physics', 'year': '1', 'credits': 12, 
     'theory': 2, 'lab': 3, 'faculty': 'Shacha Thinley'},
    {'code': 'TPH302', 'name': 'Statistical Mechanics', 'program': 'Physics', 'year': '3', 'credits': 12, 
     'theory': 4, 'lab': 0, 'faculty': 'Nachiketa Homchaudhuri'},
    {'code': 'MEC204', 'name': 'Mechanics II', 'program': 'Physics', 'year': '2', 'credits': 12, 
     'theory': 3, 'lab': 3, 'faculty': 'Nachiketa Homchaudhuri'},
    {'code': 'MEC203', 'name': 'Electromagnetism', 'program': 'Physics', 'year': '2', 'credits': 12, 
     'theory': 3, 'lab': 3, 'faculty': 'Mon Bahadur Ghalley'},
    {'code': 'MPH304', 'name': 'Quantum Mechanics', 'program': 'Physics', 'year': '3', 'credits': 12, 
     'theory': 4, 'lab': 0, 'faculty': 'Mon Bahadur Ghalley'},
    {'code': 'MMP302', 'name': 'Computational Physics', 'program': 'Physics', 'year': '3', 'credits': 12, 
     'theory': 3, 'lab': 3, 'faculty': 'Dr. Karma Tenzin'},
    {'code': 'MMP201', 'name': 'Mathematical Physics I', 'program': 'Physics', 'year': '2', 'credits': 12, 
     'theory': 4, 'lab': 0, 'faculty': 'Rit Wik Sharma'},
    {'code': 'CME101', 'name': 'Newtonian Mechanics', 'program': 'Physics', 'year': '1', 'credits': 12, 
     'theory': 2, 'lab': 3, 'faculty': 'Rit Wik Sharma'},
    {'code': 'PLT101', 'name': 'Programming Fundamentals', 'program': 'Physics', 'year': '2', 'credits': 12, 
     'theory': 3, 'lab': 0, 'faculty': ''},
    {'code': 'FMT101', 'name': 'Fundamentals of Mathematics', 'program': 'Physics', 'year': '1', 'credits': 12, 
     'theory': 3, 'lab': 0, 'faculty': ''},
    
    # Life Sciences Modules
    {'code': 'BTZ101', 'name': 'Fundamentals of Life Science', 'program': 'Life Sciences', 'year': '1', 
     'credits': 12, 'theory': 3, 'lab': 4, 'faculty': 'Ugyen Dorji Tamang'},
    {'code': 'BTS204', 'name': 'Embryology of Angiosperms', 'program': 'Life Sciences', 'year': '2', 
     'credits': 12, 'theory': 3, 'lab': 6, 'faculty': 'Karma Wangchuck'},
    {'code': 'ZLS205', 'name': 'Biochemistry', 'program': 'Life Sciences', 'year': '2', 'credits': 12, 
     'theory': 3, 'lab': 6, 'faculty': 'Mohan Singh Rana'},
    {'code': 'BTZ202', 'name': 'Microbiology', 'program': 'Life Sciences', 'year': '2', 'credits': 12, 
     'theory': 3, 'lab': 6, 'faculty': 'Ugyen Dorji Tamang'},
    {'code': 'ZLS204', 'name': 'Developmental Biology', 'program': 'Life Sciences', 'year': '2', 'credits': 12, 
     'theory': 4, 'lab': 0, 'faculty': 'Tshering Dekar'},
    {'code': 'AMT202', 'name': 'Foundation Statistics', 'program': 'Life Sciences', 'year': '2', 'credits': 12, 
     'theory': 2, 'lab': 3, 'faculty': 'Paul Raj'},
    {'code': 'BTS306', 'name': 'Plant Breeding & Horticulture', 'program': 'Life Sciences', 'year': '3', 
     'credits': 12, 'theory': 3, 'lab': 6, 'faculty': 'Bimal Kumar Chetri'},
    {'code': 'ZLS307', 'name': 'Freshwater Biology', 'program': 'Life Sciences', 'year': '3', 'credits': 12, 
     'theory': 3, 'lab': 6, 'faculty': 'Tshering Dekar'},
    {'code': 'BTS307', 'name': 'Economic Botany', 'program': 'Life Sciences', 'year': '3', 'credits': 12, 
     'theory': 3, 'lab': 6, 'faculty': 'Karma Wangchuck'},
    {'code': 'ZLS308', 'name': 'Animal Physiology', 'program': 'Life Sciences', 'year': '3', 'credits': 12, 
     'theory': 3, 'lab': 6, 'faculty': 'Mohan Singh Rana'},
    {'code': 'BTS308', 'name': 'Plant Biotechnology & Tissue Culture', 'program': 'Life Sciences', 'year': '3', 
     'credits': 12, 'theory': 3, 'lab': 3, 'faculty': 'Sonam Tobgay'},
    {'code': 'FMT101', 'name': 'Fundamentals of Mathematics', 'program': 'Life Sciences', 'year': '1', 
     'credits': 12, 'theory': 3, 'lab': 2, 'faculty': 'LS X'},
    {'code': 'ACS101', 'name': 'Academic Skills', 'program': 'Life Sciences', 'year': '1', 
     'credits': 12, 'theory': 4, 'lab': 0, 'faculty': 'LS Z'},
    {'code': 'LAC101', 'name': 'Dzongkha Shoeyon Dretshel', 'program': 'Life Sciences', 'year': '1', 
     'credits': 12, 'theory': 4, 'lab': 0, 'faculty': 'LS Y'}
]

# Workload Data - Each unique module counted as 1 (theory + lab combined)
workload_data = [
    # Chemistry Faculty Workload
    {'faculty': 'Dr. Jas Raj Subba', 'program': 'Chemistry', 'code': 'OCH201', 
     'name': 'Organic Chemistry I', 'year': '2', 'theory': 3, 'lab': 3, 'students': 30, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 2},
    {'faculty': 'Dr. Jas Raj Subba', 'program': 'Chemistry', 'code': 'FCH101', 
     'name': 'Fundamentals of Chemistry', 'year': '1', 'theory': 3, 'lab': 2, 'students': 37, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 1},
    
    {'faculty': 'Mr. Sangay Wangchuk', 'program': 'Chemistry', 'code': 'PCH303', 
     'name': 'Physical Chemistry III', 'year': '3', 'theory': 3, 'lab': 3, 'students': 32, 
     'prep': 1.2, 'assess': 'High', 'shadow': 'No', 'admin': 3},
    {'faculty': 'Mr. Sangay Wangchuk', 'program': 'Chemistry', 'code': 'RSM301', 
     'name': 'Research Methods', 'year': '2', 'theory': 4, 'lab': 0, 'students': 30, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 0},
    
    {'faculty': 'Mrs. Punam Mafchan', 'program': 'Chemistry', 'code': 'OCH304', 
     'name': 'Spectroscopic Methods in Chemistry', 'year': '3', 'theory': 4, 'lab': 0, 
     'students': 32, 'prep': 1.2, 'assess': 'Medium', 'shadow': 'No', 'admin': 0},
    {'faculty': 'Mrs. Punam Mafchan', 'program': 'Chemistry', 'code': 'PCH201', 
     'name': 'Physical Chemistry I', 'year': '2', 'theory': 3, 'lab': 3, 'students': 30, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 0},
    
    {'faculty': 'Ms. Kuenzang Choki', 'program': 'Chemistry', 'code': 'OCH303', 
     'name': 'Organic Chemistry III', 'year': '3', 'theory': 3, 'lab': 3, 'students': 32, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 0},
    {'faculty': 'Ms. Kuenzang Choki', 'program': 'Chemistry', 'code': 'FCH101', 
     'name': 'Fundamentals of Chemistry', 'year': '1', 'theory': 3, 'lab': 2, 'students': 37, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 0},
    
    {'faculty': 'Ms. Sangay Yuden', 'program': 'Chemistry', 'code': 'BCH301', 
     'name': 'Principles of Biochemistry I', 'year': '3', 'theory': 3, 'lab': 3, 'students': 32, 
     'prep': 1.0, 'assess': 'High', 'shadow': 'No', 'admin': 1},
    {'faculty': 'Ms. Sangay Yuden', 'program': 'Chemistry', 'code': 'FCH101', 
     'name': 'Fundamentals of Chemistry (Life Science)', 'year': '1', 'theory': 3, 'lab': 2, 'students': 37, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 1},
    
    {'faculty': 'Mr. Tashi Dendup', 'program': 'Chemistry', 'code': 'ICH202', 
     'name': 'Inorganic Chemistry II', 'year': '2', 'theory': 3, 'lab': 3, 'students': 30, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 0},
    {'faculty': 'Mr. Tashi Dendup', 'program': 'Chemistry', 'code': 'ICH203', 
     'name': 'Inorganic Chemistry III', 'year': '3', 'theory': 3, 'lab': 3, 'students': 32, 
     'prep': 1.2, 'assess': 'High', 'shadow': 'No', 'admin': 1},
    
    {'faculty': 'Dechen Lhendup', 'program': 'Chemistry', 'code': 'FMT204', 
     'name': 'Foundations of Mathematics III', 'year': '2', 'theory': 5, 'lab': 0, 'students': 30, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 0},
    {'faculty': 'DS-Y', 'program': 'Chemistry', 'code': 'CSP101', 
     'name': 'Foundations of Python Programming', 'year': '1', 'theory': 3, 'lab': 2, 'students': 37, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 0},
    {'faculty': 'S. Chitra', 'program': 'Chemistry', 'code': 'ACS101', 
     'name': 'Academic Skills', 'year': '1', 'theory': 4, 'lab': 0, 'students': 37, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 0},
    
    # Physics Faculty Workload
    {'faculty': 'Shacha Thinley', 'program': 'Physics', 'code': 'MPH303', 
     'name': 'Atomic Physics', 'year': '3', 'theory': 3, 'lab': 3, 'students': 35, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 2},
    {'faculty': 'Shacha Thinley', 'program': 'Physics', 'code': 'PHY101', 
     'name': 'Fundamentals of Physics (Chemistry)', 'year': '1', 'theory': 2, 'lab': 3, 'students': 36, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 1},
    
    {'faculty': 'Nachiketa Homchaudhuri', 'program': 'Physics', 'code': 'TPH302', 
     'name': 'Statistical Mechanics', 'year': '3', 'theory': 4, 'lab': 0, 'students': 33, 
     'prep': 1.2, 'assess': 'High', 'shadow': 'No', 'admin': 4},
    {'faculty': 'Nachiketa Homchaudhuri', 'program': 'Physics', 'code': 'MEC204', 
     'name': 'Mechanics II', 'year': '2', 'theory': 3, 'lab': 3, 'students': 28, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 2},
    
    {'faculty': 'Mon Bahadur Ghalley', 'program': 'Physics', 'code': 'MEC203', 
     'name': 'Electromagnetism', 'year': '2', 'theory': 3, 'lab': 3, 'students': 28, 
     'prep': 1.0, 'assess': 'High', 'shadow': 'No', 'admin': 2},
    {'faculty': 'Mon Bahadur Ghalley', 'program': 'Physics', 'code': 'MPH304', 
     'name': 'Quantum Mechanics', 'year': '3', 'theory': 4, 'lab': 0, 'students': 33, 
     'prep': 1.2, 'assess': 'High', 'shadow': 'No', 'admin': 3},
    
    {'faculty': 'Dr. Karma Tenzin', 'program': 'Physics', 'code': 'MMP302', 
     'name': 'Computational Physics', 'year': '3', 'theory': 3, 'lab': 3, 'students': 33, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 2},
    {'faculty': 'Dr. Karma Tenzin', 'program': 'Physics', 'code': 'RSM301', 
     'name': 'Research Methods (Physics)', 'year': '2', 'theory': 4, 'lab': 0, 'students': 33, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 2},
    
    {'faculty': 'Rit Wik Sharma', 'program': 'Physics', 'code': 'MMP201', 
     'name': 'Mathematical Physics I', 'year': '2', 'theory': 4, 'lab': 0, 'students': 28, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 2},
    {'faculty': 'Rit Wik Sharma', 'program': 'Physics', 'code': 'CME101', 
     'name': 'Newtonian Mechanics', 'year': '1', 'theory': 2, 'lab': 3, 'students': 33, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 1},
    
    # Life Sciences Core Faculty Workload - Each module counted once
    {'faculty': 'Ugyen Dorji Tamang', 'program': 'Life Sciences', 'code': 'BTZ101', 
     'name': 'Fundamentals of Life Science', 'year': '1', 'theory': 3, 'lab': 4, 'students': 42, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'Yes', 'admin': 2},
    {'faculty': 'Ugyen Dorji Tamang', 'program': 'Life Sciences', 'code': 'BTZ202', 
     'name': 'Microbiology', 'year': '2', 'theory': 3, 'lab': 6, 'students': 33, 
     'prep': 1.0, 'assess': 'High', 'shadow': 'Yes', 'admin': 2},
    
    {'faculty': 'Karma Wangchuck', 'program': 'Life Sciences', 'code': 'BTS204', 
     'name': 'Embryology of Angiosperms', 'year': '2', 'theory': 3, 'lab': 6, 'students': 33, 
     'prep': 1.0, 'assess': 'High', 'shadow': 'Yes', 'admin': 2},
    {'faculty': 'Karma Wangchuck', 'program': 'Life Sciences', 'code': 'BTS307', 
     'name': 'Economic Botany', 'year': '3', 'theory': 3, 'lab': 6, 'students': 22, 
     'prep': 1.2, 'assess': 'High', 'shadow': 'No', 'admin': 2},
    
    {'faculty': 'Mohan Singh Rana', 'program': 'Life Sciences', 'code': 'ZLS205', 
     'name': 'Biochemistry', 'year': '2', 'theory': 3, 'lab': 6, 'students': 33, 
     'prep': 1.0, 'assess': 'High', 'shadow': 'Yes', 'admin': 2},
    {'faculty': 'Mohan Singh Rana', 'program': 'Life Sciences', 'code': 'ZLS308', 
     'name': 'Animal Physiology', 'year': '3', 'theory': 3, 'lab': 6, 'students': 22, 
     'prep': 1.2, 'assess': 'High', 'shadow': 'No', 'admin': 2},
    
    {'faculty': 'Tshering Dekar', 'program': 'Life Sciences', 'code': 'ZLS204', 
     'name': 'Developmental Biology', 'year': '2', 'theory': 4, 'lab': 0, 'students': 33, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 2},
    {'faculty': 'Tshering Dekar', 'program': 'Life Sciences', 'code': 'ZLS307', 
     'name': 'Freshwater Biology', 'year': '3', 'theory': 3, 'lab': 6, 'students': 22, 
     'prep': 1.2, 'assess': 'High', 'shadow': 'No', 'admin': 2},
    
    {'faculty': 'Bimal Kumar Chetri', 'program': 'Life Sciences', 'code': 'BTS306', 
     'name': 'Plant Breeding & Horticulture', 'year': '3', 'theory': 3, 'lab': 6, 'students': 22, 
     'prep': 1.2, 'assess': 'High', 'shadow': 'No', 'admin': 2},
    
    {'faculty': 'Sonam Tobgay', 'program': 'Life Sciences', 'code': 'BTS308', 
     'name': 'Plant Biotechnology & Tissue Culture', 'year': '3', 'theory': 3, 'lab': 3, 'students': 22, 
     'prep': 1.2, 'assess': 'High', 'shadow': 'No', 'admin': 3},
    
    # Service Faculty from DSDA
    {'faculty': 'Paul Raj', 'program': 'Life Sciences', 'code': 'AMT202', 
     'name': 'Foundation Statistics', 'year': '2', 'theory': 2, 'lab': 3, 'students': 33, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 0},
    {'faculty': 'LS W', 'program': 'Life Sciences', 'code': 'FCH101', 
     'name': 'Fundamentals of Chemistry (Life Science)', 'year': '1', 'theory': 2, 'lab': 2, 'students': 42, 
     'prep': 1.0, 'assess': 'Low', 'shadow': 'No', 'admin': 0},
    {'faculty': 'LS X', 'program': 'Life Sciences', 'code': 'FMT101', 
     'name': 'Fundamentals of Mathematics', 'year': '1', 'theory': 3, 'lab': 2, 'students': 42, 
     'prep': 1.0, 'assess': 'Low', 'shadow': 'No', 'admin': 0},
    {'faculty': 'LS Y', 'program': 'Life Sciences', 'code': 'LAC101', 
     'name': 'Dzongkha Shoeyon Dretshel', 'year': '1', 'theory': 4, 'lab': 0, 'students': 42, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 0},
    {'faculty': 'LS Z', 'program': 'Life Sciences', 'code': 'ACS101', 
     'name': 'Academic Skills', 'year': '1', 'theory': 4, 'lab': 0, 'students': 42, 
     'prep': 1.0, 'assess': 'Medium', 'shadow': 'No', 'admin': 0}
]

# ============================================================
# WAM CALCULATION ENGINE
# ============================================================

def calculate_wam(faculty_name, workload_data):
    """Calculate WAM score for a faculty member - each module counted once"""
    total_wam = 0
    assess_score = {'Low': 2, 'Medium': 4, 'High': 7}
    
    assignments = [w for w in workload_data if w['faculty'] == faculty_name]
    
    for work in assignments:
        theory_hrs = float(work.get('theory', 0) or 0)
        lab_hrs = float(work.get('lab', 0) or 0)
        prep_mult = float(work.get('prep', 1.0) or 1.0)
        admin_hrs = float(work.get('admin', 0) or 0)
        students = int(work.get('students', 0) or 0)
        
        base_load = (theory_hrs * prep_mult) + lab_hrs + admin_hrs
        intensity = assess_score.get(work.get('assess', 'Medium'), 4)
        student_loading = (students * intensity) * 0.04
        
        if work.get('shadow') == 'Yes':
            student_loading = student_loading * 0.5
        
        total_wam += base_load + student_loading
    
    return round(total_wam, 1)

def get_status(wam, module_count=None):
    """Get status based on WAM score - 2 modules with 35 students is balanced"""
    if wam == 0:
        return 'Empty'
    # With 2 modules at 35 students = Balanced (around 10-19 WAM)
    if wam < 10:
        return 'Sharing'
    if wam <= 19:
        return 'Balanced'
    return 'Review'

# ============================================================
# EXCEL EXPORTER
# ============================================================

def export_to_excel():
    """Export all data to Excel file with 5 sheets"""
    
    print("=" * 70)
    print("DEPARTMENT OF NATURAL SCIENCES - WORKLOAD REPORT GENERATOR")
    print("Sherubtse College - Autumn 2026")
    print("=" * 70)
    print("\n📊 Generating Excel report...")
    
    timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
    filename = f"NaturalSciences_Workload_Autumn2026_{timestamp}.xlsx"
    
    wb = Workbook()
    
    # Color definitions for programs
    program_colors = {
        'Chemistry': 'FFE4B5',      # Light goldenrod
        'Physics': 'B0D4F1',        # Light blue
        'Life Sciences': 'C1E1C1'   # Light green
    }
    
    # Status colors
    status_colors = {
        'Balanced': '92D050',       # Green
        'Sharing': 'FFC000',    # Yellow
        'Review': 'FF6B6B',     # Red
        'Empty': 'D9D9D9'           # Gray
    }
    
    # Define border style
    thin_border = Border(
        left=Side(style='thin'),
        right=Side(style='thin'),
        top=Side(style='thin'),
        bottom=Side(style='thin')
    )
    
    # ============================================================
    # SHEET 1: DASHBOARD
    # ============================================================
    print("   ✓ Creating Dashboard...")
    ws1 = wb.active
    ws1.title = "Dashboard"
    
    # Header
    ws1['A1'] = "Department of Natural Sciences - Workload Management Dashboard"
    ws1['A1'].font = Font(size=16, bold=True, color="FFFFFF")
    ws1['A1'].fill = PatternFill(start_color="7A1A1A", end_color="7A1A1A", fill_type="solid")
    ws1.merge_cells('A1:H1')
    
    ws1['A2'] = f"Report Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}"
    ws1['A2'].font = Font(size=11, italic=True)
    ws1.merge_cells('A2:H2')
    
    ws1['A3'] = "Semester: Autumn 2026 (July - December) | New Department - Merged Programmes"
    ws1['A3'].font = Font(size=11, bold=True)
    ws1.merge_cells('A3:H3')
    
    # Statistics Section
    ws1['A5'] = "DEPARTMENT STATISTICS"
    ws1['A5'].font = Font(size=14, bold=True, color="7A1A1A")
    ws1.merge_cells('A5:F5')
    
    # Calculate statistics
    core_faculty = [f for f in faculty_data if f.get('core', False)]
    service_faculty = [f for f in faculty_data if not f.get('core', True)]
    total_faculty = len(faculty_data)
    total_modules = len(workload_data)
    
    chemistry_faculty = len([f for f in faculty_data if f['program'] == 'Chemistry' and f.get('core', False)])
    physics_faculty = len([f for f in faculty_data if f['program'] == 'Physics' and f.get('core', False)])
    life_sciences_faculty = len([f for f in faculty_data if f['program'] == 'Life Sciences' and f.get('core', False)])
    
    chemistry_modules = len([w for w in workload_data if w['program'] == 'Chemistry'])
    physics_modules = len([w for w in workload_data if w['program'] == 'Physics'])
    life_sciences_modules = len([w for w in workload_data if w['program'] == 'Life Sciences'])
    
    # Calculate WAM stats
    wam_scores = []
    balanced = Sharing = Review = 0
    
    for faculty in faculty_data:
        wam = calculate_wam(faculty['name'], workload_data)
        if wam > 0:
            wam_scores.append(wam)
        status = get_status(wam)
        if status == 'Balanced':
            balanced += 1
        elif status == 'Review':
            Review += 1
        elif status == 'Sharing':
            Sharing += 1
    
    avg_wam = sum(wam_scores) / len(wam_scores) if wam_scores else 0
    
    stats_data = [
        ['Total Faculty (All)', total_faculty],
        ['Core Faculty', len(core_faculty)],
        ['Service Faculty (DSDA)', len(service_faculty)],
        ['Total Modules (Unique)', total_modules],
        ['Chemistry Core Faculty', chemistry_faculty],
        ['Physics Core Faculty', physics_faculty],
        ['Life Sciences Core Faculty', life_sciences_faculty],
        ['Chemistry Modules', chemistry_modules],
        ['Physics Modules', physics_modules],
        ['Life Sciences Modules', life_sciences_modules],
        ['Average WAM', f"{avg_wam:.1f}"],
        ['Balanced Faculty (Green)', balanced],
        ['Sharing Faculty (Yellow)', Sharing],
        ['Review Faculty (Red)', Review]
    ]
    
    row = 7
    for i, (label, value) in enumerate(stats_data):
        ws1[f'A{row+i}'] = label
        ws1[f'A{row+i}'].font = Font(bold=True)
        ws1[f'B{row+i}'] = value
        ws1[f'B{row+i}'].font = Font(size=12, bold=True)
        
        # Color coding for status
        if label == 'Balanced Faculty (Green)':
            ws1[f'B{row+i}'].fill = PatternFill(start_color="92D050", end_color="92D050", fill_type="solid")
        elif label == 'Sharing Faculty (Yellow)':
            ws1[f'B{row+i}'].fill = PatternFill(start_color="FFC000", end_color="FFC000", fill_type="solid")
        elif label == 'Review Faculty (Red)':
            ws1[f'B{row+i}'].fill = PatternFill(start_color="FF6B6B", end_color="FF6B6B", fill_type="solid")
    
    # Program Summary Table
    ws1['A24'] = "PROGRAM SUMMARIES"
    ws1['A24'].font = Font(size=14, bold=True, color="7A1A1A")
    ws1.merge_cells('A24:E24')
    
    prog_headers = ['Program', 'Core Faculty', 'Modules', 'Avg WAM', 'Status']
    for col, header in enumerate(prog_headers, start=1):
        cell = ws1.cell(row=25, column=col, value=header)
        cell.font = Font(bold=True, color="FFFFFF")
        cell.fill = PatternFill(start_color="7A1A1A", end_color="7A1A1A", fill_type="solid")
        cell.alignment = Alignment(horizontal='center')
        cell.border = thin_border
    
    programs = ['Chemistry', 'Physics', 'Life Sciences']
    row = 26
    for prog in programs:
        faculty_in_prog = [f for f in faculty_data if f['program'] == prog and f.get('core', False)]
        modules_in_prog = [w for w in workload_data if w['program'] == prog]
        
        wam_sum = 0
        for f in faculty_in_prog:
            wam_sum += calculate_wam(f['name'], workload_data)
        avg = wam_sum / len(faculty_in_prog) if faculty_in_prog else 0
        
        if avg > 19:
            status = "🔴 Review"
            status_color = "FF6B6B"
        elif avg >= 10:
            status = "✅ Balanced"
            status_color = "92D050"
        else:
            status = "🟡 Monitor"
            status_color = "FFC000"
        
        ws1[f'A{row}'] = f"BSc in {prog}"
        ws1[f'B{row}'] = len(faculty_in_prog)
        ws1[f'C{row}'] = len(modules_in_prog)
        ws1[f'D{row}'] = f"{avg:.1f}"
        ws1[f'E{row}'] = status
        
        # Color the program row
        for col in range(1, 6):
            cell = ws1.cell(row=row, column=col)
            cell.fill = PatternFill(start_color=program_colors.get(prog, 'FFFFFF'), 
                                   end_color=program_colors.get(prog, 'FFFFFF'), 
                                   fill_type="solid")
            cell.border = thin_border
        
        row += 1
    
    # Faculty WAM List
    ws1['A32'] = "FACULTY WAM SUMMARY"
    ws1['A32'].font = Font(size=14, bold=True, color="7A1A1A")
    ws1.merge_cells('A32:E32')
    
    faculty_headers = ['Faculty Name', 'Program', 'Type', 'Modules', 'WAM Score', 'Status']
    for col, header in enumerate(faculty_headers, start=1):
        cell = ws1.cell(row=33, column=col, value=header)
        cell.font = Font(bold=True, color="FFFFFF")
        cell.fill = PatternFill(start_color="7A1A1A", end_color="7A1A1A", fill_type="solid")
        cell.alignment = Alignment(horizontal='center')
        cell.border = thin_border
    
    row = 34
    for faculty in faculty_data:
        wam = calculate_wam(faculty['name'], workload_data)
        status = get_status(wam)
        faculty_type = 'Core' if faculty.get('core', False) else 'Service (DSDA)'
        module_count = len([w for w in workload_data if w['faculty'] == faculty['name']])
        
        ws1[f'A{row}'] = faculty['name']
        ws1[f'B{row}'] = faculty['program']
        ws1[f'C{row}'] = faculty_type
        ws1[f'D{row}'] = module_count
        ws1[f'E{row}'] = wam
        ws1[f'F{row}'] = status
        
        # Color status
        status_color = status_colors.get(status, 'D9D9D9')
        for col in range(1, 7):
            cell = ws1.cell(row=row, column=col)
            cell.border = thin_border
            if col == 6:
                cell.fill = PatternFill(start_color=status_color, end_color=status_color, fill_type="solid")
                cell.font = Font(bold=True)
        
        # Color program column
        prog_color = program_colors.get(faculty['program'], 'FFFFFF')
        ws1[f'B{row}'].fill = PatternFill(start_color=prog_color, end_color=prog_color, fill_type="solid")
        
        row += 1
    
    # Column widths
    ws1.column_dimensions['A'].width = 28
    ws1.column_dimensions['B'].width = 18
    ws1.column_dimensions['C'].width = 18
    ws1.column_dimensions['D'].width = 12
    ws1.column_dimensions['E'].width = 12
    ws1.column_dimensions['F'].width = 15
    
    # ============================================================
    # SHEET 2: AUTUMN 2026 WORKLOAD
    # ============================================================
    print("   ✓ Creating Autumn 2026 Workload...")
    ws2 = wb.create_sheet("Autumn 2026 Workload")
    
    workload_headers = ['Faculty', 'Program', 'Type', 'Module Code', 'Module Name', 'Year', 
                       'Theory Hrs', 'Lab Hrs', 'Total Hrs', 'Students', 'Prep Mult', 'Assessment', 
                       'Shadow Support', 'Admin Hrs']
    
    for col, header in enumerate(workload_headers, start=1):
        cell = ws2.cell(row=1, column=col, value=header)
        cell.font = Font(bold=True, color="FFFFFF")
        cell.fill = PatternFill(start_color="7A1A1A", end_color="7A1A1A", fill_type="solid")
        cell.alignment = Alignment(horizontal='center')
        cell.border = thin_border
    
    row = 2
    for work in workload_data:
        # Get faculty type
        faculty_obj = next((f for f in faculty_data if f['name'] == work['faculty']), None)
        faculty_type = 'Core' if faculty_obj and faculty_obj.get('core', False) else 'Service (DSDA)'
        total_hrs = float(work.get('theory', 0) or 0) + float(work.get('lab', 0) or 0)
        
        data_row = [
            work['faculty'],
            work['program'],
            faculty_type,
            work['code'],
            work['name'],
            work['year'],
            work['theory'],
            work['lab'],
            total_hrs,
            work['students'],
            work['prep'],
            work['assess'],
            work['shadow'],
            work['admin']
        ]
        
        for col_idx, value in enumerate(data_row, start=1):
            cell = ws2.cell(row=row, column=col_idx, value=value)
            cell.border = thin_border
            
            # Color by program (column 2)
            if col_idx == 2:
                color = program_colors.get(work['program'], 'FFFFFF')
                cell.fill = PatternFill(start_color=color, end_color=color, fill_type="solid")
        
        row += 1
    
    # Column widths
    ws2.column_dimensions['A'].width = 25
    ws2.column_dimensions['B'].width = 18
    ws2.column_dimensions['C'].width = 18
    ws2.column_dimensions['D'].width = 15
    ws2.column_dimensions['E'].width = 35
    ws2.column_dimensions['F'].width = 10
    ws2.column_dimensions['G'].width = 12
    ws2.column_dimensions['H'].width = 12
    ws2.column_dimensions['I'].width = 12
    ws2.column_dimensions['J'].width = 12
    ws2.column_dimensions['K'].width = 12
    ws2.column_dimensions['L'].width = 15
    ws2.column_dimensions['M'].width = 15
    ws2.column_dimensions['N'].width = 12
    
    # ============================================================
    # SHEET 3: MODULES MASTER LIST
    # ============================================================
    print("   ✓ Creating Modules Master List...")
    ws3 = wb.create_sheet("Modules Master List")
    
    module_headers = ['Code', 'Module Name', 'Program', 'Year', 'Credits', 
                     'Theory Hrs/Wk', 'Lab Hrs/Wk', 'Total Hrs', 'Default Faculty']
    
    for col, header in enumerate(module_headers, start=1):
        cell = ws3.cell(row=1, column=col, value=header)
        cell.font = Font(bold=True, color="FFFFFF")
        cell.fill = PatternFill(start_color="7A1A1A", end_color="7A1A1A", fill_type="solid")
        cell.alignment = Alignment(horizontal='center')
        cell.border = thin_border
    
    row = 2
    for module in module_data:
        total_hrs = float(module.get('theory', 0) or 0) + float(module.get('lab', 0) or 0)
        data_row = [
            module.get('code', ''),
            module.get('name', ''),
            module.get('program', ''),
            module.get('year', ''),
            module.get('credits', ''),
            module.get('theory', ''),
            module.get('lab', ''),
            total_hrs,
            module.get('faculty', '')
        ]
        
        for col_idx, value in enumerate(data_row, start=1):
            cell = ws3.cell(row=row, column=col_idx, value=value)
            cell.border = thin_border
            
            # Color by program (column 3)
            if col_idx == 3:
                color = program_colors.get(module.get('program', ''), 'FFFFFF')
                cell.fill = PatternFill(start_color=color, end_color=color, fill_type="solid")
        
        row += 1
    
    # Column widths
    ws3.column_dimensions['A'].width = 15
    ws3.column_dimensions['B'].width = 35
    ws3.column_dimensions['C'].width = 18
    ws3.column_dimensions['D'].width = 10
    ws3.column_dimensions['E'].width = 12
    ws3.column_dimensions['F'].width = 15
    ws3.column_dimensions['G'].width = 15
    ws3.column_dimensions['H'].width = 12
    ws3.column_dimensions['I'].width = 25
    
    # ============================================================
    # SHEET 4: WAM CALCULATOR (PRE-POPULATED)
    # ============================================================
    print("   ✓ Creating WAM Calculator...")
    ws4 = wb.create_sheet("WAM Calculator")

    ws4['A1'] = "WORKLOAD ASSIGNMENT WAM CALCULATOR"
    ws4['A1'].font = Font(size=16, bold=True, color="7A1A1A")
    ws4.merge_cells('A1:L1')

    ws4['A2'] = "Enter workload details below to calculate WAM contributions"
    ws4['A2'].font = Font(size=11, italic=True)
    ws4.merge_cells('A2:L2')

    ws4['A3'] = "📊 WAM Thresholds: 0-9 = Sharing (Yellow) | 10-19 = Balanced (Green) | 20+ = Review (Red)"
    ws4['A3'].font = Font(size=10, bold=True, color="7A1A1A")
    ws4.merge_cells('A3:L3')

    wam_headers = ['Faculty', 'Program', 'Module Code', 'Module Name', 'Year', 
                  'Theory Hrs', 'Lab Hrs', 'Total Hrs', 'Students', 'Prep Mult', 'Assessment', 
                  'Shadow Support', 'Admin Hrs']

    for col, header in enumerate(wam_headers, start=1):
        cell = ws4.cell(row=5, column=col, value=header)
        cell.font = Font(bold=True, color="FFFFFF")
        cell.fill = PatternFill(start_color="7A1A1A", end_color="7A1A1A", fill_type="solid")
        cell.alignment = Alignment(horizontal='center')
        cell.border = thin_border

    # POPULATE WITH ACTUAL WORKLOAD DATA
    row = 6
    for work in workload_data:
        total_hrs = float(work.get('theory', 0) or 0) + float(work.get('lab', 0) or 0)
        
        data_row = [
            work['faculty'],
            work['program'],
            work['code'],
            work['name'],
            work['year'],
            work['theory'],
            work['lab'],
            total_hrs,
            work['students'],
            work['prep'],
            work['assess'],
            work['shadow'],
            work['admin']
        ]
        
        for col_idx, value in enumerate(data_row, start=1):
            cell = ws4.cell(row=row, column=col_idx, value=value)
            cell.border = thin_border
            
            # Color by program (column 2)
            if col_idx == 2:
                color = program_colors.get(work['program'], 'FFFFFF')
                cell.fill = PatternFill(start_color=color, end_color=color, fill_type="solid")
        
        row += 1

    # Add some empty rows at the end for additional entries
    empty_start = row
    for i in range(10):  # Add 10 empty rows
        for col in range(1, 14):
            cell = ws4.cell(row=empty_start + i, column=col, value='')
            cell.border = thin_border
    
    # Add notes
    notes_row = empty_start + 12
    ws4['A' + str(notes_row)] = "📝 INSTRUCTIONS:"
    ws4['A' + str(notes_row)].font = Font(bold=True)
    ws4['B' + str(notes_row)] = "1. Each module should be entered ONCE (theory + lab combined)"
    ws4['B' + str(notes_row + 1)] = "2. Enter your name and program"
    ws4['B' + str(notes_row + 2)] = "3. Fill in theory/lab hours, student numbers, and assessment intensity"
    ws4['B' + str(notes_row + 3)] = "4. The WAM will be calculated automatically using:"
    ws4['B' + str(notes_row + 4)] = "   WAM = (Theory × Prep) + Lab + Admin + (Students × Assessment × 0.04)"
    ws4['B' + str(notes_row + 5)] = "5. Assessment Intensity: Low=2, Medium=4, High=7"
    ws4['B' + str(notes_row + 6)] = "6. Prep Multiplier: 0.8=Established, 1.0=Standard, 1.2=New/Complex"
    
    # Column widths
    for col in range(1, 14):
        ws4.column_dimensions[get_column_letter(col)].width = 15
    ws4.column_dimensions['A'].width = 25
    ws4.column_dimensions['D'].width = 35
    ws4.column_dimensions['B'].width = 18

    # ============================================================
    # SHEET 5: DATA ENTRY - FUTURE PLANS
    # ============================================================
    print("   ✓ Creating Data Entry - Future Plans...")
    ws5 = wb.create_sheet("Data Entry - Future Plans")
    
    ws5['A1'] = "FUTURE PLANNING - DATA ENTRY WORKSHEET"
    ws5['A1'].font = Font(size=16, bold=True, color="7A1A1A")
    ws5.merge_cells('A1:M1')
    
    ws5['A2'] = "Department of Natural Sciences - Inclusive Planning for Autumn 2026 and Beyond"
    ws5['A2'].font = Font(size=11, italic=True)
    ws5.merge_cells('A2:M2')
    
    ws5['A3'] = "📌 All faculty members are invited to indicate their preferred module allocations"
    ws5['A3'].font = Font(size=10, bold=True, color="7A1A1A")
    ws5.merge_cells('A3:M3')
    
    planning_headers = ['Faculty Name', 'Program', 'Designation', 'Type', 'Current Modules', 
                       'Current WAM', 'Status', 'Specialization', 
                       'Module 1 (Comfortable)', 'Module 2 (Comfortable)', 'Module 3 (Comfortable)',
                       'Module 4 (Optional)', 'Notes']
    
    for col, header in enumerate(planning_headers, start=1):
        cell = ws5.cell(row=6, column=col, value=header)
        cell.font = Font(bold=True, color="FFFFFF")
        cell.fill = PatternFill(start_color="7A1A1A", end_color="7A1A1A", fill_type="solid")
        cell.alignment = Alignment(horizontal='center', wrap_text=True)
        cell.border = thin_border
    
    row = 7
    for faculty in faculty_data:
        wam = calculate_wam(faculty['name'], workload_data)
        status = get_status(wam)
        faculty_type = 'Core' if faculty.get('core', False) else 'Service (DSDA)'
        module_count = len([w for w in workload_data if w['faculty'] == faculty['name']])
        
        status_color = status_colors.get(status, 'D9D9D9')
        
        ws5[f'A{row}'] = faculty['name']
        ws5[f'B{row}'] = faculty['program']
        ws5[f'C{row}'] = faculty['designation']
        ws5[f'D{row}'] = faculty_type
        ws5[f'E{row}'] = module_count
        ws5[f'F{row}'] = wam
        ws5[f'G{row}'] = status
        ws5[f'H{row}'] = faculty['specialization']
        ws5[f'I{row}'] = ''  # Module 1 - Empty
        ws5[f'J{row}'] = ''  # Module 2 - Empty
        ws5[f'K{row}'] = ''  # Module 3 - Empty
        ws5[f'L{row}'] = ''  # Module 4 - Empty
        ws5[f'M{row}'] = 'Please indicate your preferred modules'  # Notes
        
        for col in range(1, 14):
            cell = ws5.cell(row=row, column=col)
            cell.border = thin_border
            
            if col == 2:
                color = program_colors.get(faculty['program'], 'FFFFFF')
                cell.fill = PatternFill(start_color=color, end_color=color, fill_type="solid")
            
            if col == 7:
                cell.fill = PatternFill(start_color=status_color, end_color=status_color, fill_type="solid")
                cell.font = Font(bold=True)
        
        row += 1
    
    # Section 2: Module Proposals
    ws5['A' + str(row + 2)] = "MODULE PROPOSALS - FOR FUTURE SEMESTERS"
    ws5['A' + str(row + 2)].font = Font(size=14, bold=True, color="7A1A1A")
    ws5.merge_cells('A' + str(row + 2) + ':I' + str(row + 2))
    
    proposal_headers = ['Module Code', 'Module Name', 'Program', 'Year', 'Credits', 
                       'Theory Hrs', 'Lab Hrs', 'Proposed Faculty', 'Status']
    
    start_row = row + 3
    for col, header in enumerate(proposal_headers, start=1):
        cell = ws5.cell(row=start_row, column=col, value=header)
        cell.font = Font(bold=True, color="FFFFFF")
        cell.fill = PatternFill(start_color="7A1A1A", end_color="7A1A1A", fill_type="solid")
        cell.alignment = Alignment(horizontal='center')
        cell.border = thin_border
    
    proposals = [
        ['', '', '', '', '', '', '', '', ''],
        ['', '', '', '', '', '', '', '', ''],
        ['', '', '', '', '', '', '', '', ''],
        ['', '', '', '', '', '', '', '', ''],
        ['', '', '', '', '', '', '', '', '']
    ]
    
    row = start_row + 1
    for prop in proposals:
        for col_idx, value in enumerate(prop, start=1):
            cell = ws5.cell(row=row, column=col_idx, value=value)
            cell.border = thin_border
        row += 1
    
    # Section 3: Summary
    ws5['A' + str(row + 2)] = "PLANNING SUMMARY"
    ws5['A' + str(row + 2)].font = Font(size=14, bold=True, color="7A1A1A")
    ws5.merge_cells('A' + str(row + 2) + ':D' + str(row + 2))
    
    summary_row = row + 3
    summary_data = [
        ['Total Core Faculty', len([f for f in faculty_data if f.get('core', False)])],
        ['Total Service Faculty (DSDA)', len([f for f in faculty_data if not f.get('core', True)])],
        ['Total Unique Modules', len(workload_data)],
        ['Balanced Faculty (10-19 WAM)', len([f for f in faculty_data if get_status(calculate_wam(f['name'], workload_data)) == 'Balanced'])],
        ['Sharing Faculty (<10 WAM)', len([f for f in faculty_data if get_status(calculate_wam(f['name'], workload_data)) == 'Sharing'])],
        ['Review Faculty (>19 WAM)', len([f for f in faculty_data if get_status(calculate_wam(f['name'], workload_data)) == 'Review'])]
    ]
    
    for i, (label, value) in enumerate(summary_data):
        ws5[f'A{summary_row + i}'] = label
        ws5[f'A{summary_row + i}'].font = Font(bold=True)
        ws5[f'B{summary_row + i}'] = value
        ws5[f'B{summary_row + i}'].font = Font(size=12, bold=True)
    
    # Column widths
    ws5.column_dimensions['A'].width = 28
    ws5.column_dimensions['B'].width = 18
    ws5.column_dimensions['C'].width = 20
    ws5.column_dimensions['D'].width = 18
    ws5.column_dimensions['E'].width = 14
    ws5.column_dimensions['F'].width = 14
    ws5.column_dimensions['G'].width = 14
    ws5.column_dimensions['H'].width = 35
    ws5.column_dimensions['I'].width = 20
    ws5.column_dimensions['J'].width = 20
    ws5.column_dimensions['K'].width = 20
    ws5.column_dimensions['L'].width = 20
    ws5.column_dimensions['M'].width = 30
    
    # ============================================================
    # SAVE THE FILE
    # ============================================================
    cwd = os.getcwd()
    full_path = os.path.join(cwd, filename)
    wb.save(full_path)
    
    file_size = os.path.getsize(full_path)
    size_kb = file_size / 1024
    
    print(f"\n✅ SUCCESS! Excel file exported to:")
    print(f"   📁 {full_path}")
    print(f"   📊 File size: {size_kb:.1f} KB")
    print(f"\n📋 The report contains 5 sheets:")
    print("   1. Dashboard - Department overview with Core/Service faculty")
    print("   2. Autumn 2026 Workload - Complete workload (each module counted once)")
    print("   3. Modules Master List - All unique modules with total hours")
    print("   4. WAM Calculator - Template for faculty to fill their comfortable modules")
    print("   5. Data Entry - Future Plans - Self-nomination form for all faculty")
    print(f"\n📈 Total Unique Modules: {len(workload_data)}")
    print(f"👨‍🏫 Core Faculty: {len([f for f in faculty_data if f.get('core', False)])}")
    print(f"👨‍🏫 Service Faculty (DSDA): {len([f for f in faculty_data if not f.get('core', True)])}")
    print("\n📊 WAM Thresholds:")
    print("   🟡 0-9: Sharing (Monitor)")
    print("   🟢 10-19: Balanced (Optimal - 2 modules with 35 students)")
    print("   🔴 20+: Review (Review)")
    print("\n💡 Note: Each module is counted as 1 unit even with separate lab components")
    print("\n" + "=" * 70)
    
    return full_path

# ============================================================
# MAIN EXECUTION
# ============================================================

if __name__ == "__main__":
    try:
        export_to_excel()
    except Exception as e:
        print(f"\n❌ ERROR: {str(e)}")
        print("\nPlease make sure you have the required packages installed:")
        print("   pip install pandas openpyxl")
```
